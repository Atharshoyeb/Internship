```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the Naukri.com website
driver.get("https://www.naukri.com/")

# Wait for the page to load
time.sleep(10)

# Enter "Data Scientist" in the search field
search_field = driver.find_element(By.XPATH, "//input[@placeholder='Skills, Designations, Companies']")
search_field.send_keys("Data Scientist")

# Click the search button
search_button = driver.find_element(By.XPATH, "//div[@class='qsbSubmit']")
search_button.click()

# Wait for the search results to load
time.sleep(10)

# Apply location filter for "Delhi/NCR"
try:
    location_filter = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//span[text()='Delhi / NCR']/preceding-sibling::i"))
    )
    location_filter.click()
except Exception as e:
    print(f"Error applying location filter: {e}")

# Apply salary filter for "3-6 lakhs"
try:
    salary_filter = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//span[text()='3-6 Lakhs']/preceding-sibling::i"))
    )
    salary_filter.click()
except Exception as e:
    print(f"Error applying salary filter: {e}")

# Wait for the filters to be applied and results to be updated
time.sleep(10)

# Scrape the first 10 job results
jobs = []
for i in range(1, 11):
    try:
        job_title = driver.find_element(By.XPATH, f"(//a[@class='title fw500 ellipsis'])[{i}]").text
        job_location = driver.find_element(By.XPATH, f"(//li[@class='fleft grey-text br2 placeHolderLi location']/span)[{i}]").text
        company_name = driver.find_element(By.XPATH, f"(//a[@class='subTitle ellipsis fleft'])[{i}]").text
        experience = driver.find_element(By.XPATH, f"(//li[@class='fleft grey-text br2 placeHolderLi experience']/span)[{i}]").text
        jobs.append([job_title, job_location, company_name, experience])
    except Exception as e:
        print(f"Error while scraping job {i}: {e}")

# Create a DataFrame
df = pd.DataFrame(jobs, columns=['Job Title', 'Location', 'Company Name', 'Experience Required'])
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the Flipkart iPhone 11 reviews page
url = "https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART"
driver.get(url)

# Close the login popup if it appears
try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), '✕')]"))
    )
    close_button.click()
except Exception as e:
    print("Login popup did not appear or could not be closed:", e)

reviews = []

# Function to scrape data from the current page
def scrape_page():
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    review_blocks = soup.find_all('div', {'class': '_1AtVbE'})
    
    for block in review_blocks:
        try:
            rating = block.find('div', {'class': '_3LWZlK'}).text
            summary = block.find('p', {'class': '_2-N8zT'}).text
            full_review = block.find('div', {'class': 't-ZTKy'}).text
            reviews.append([rating, summary, full_review])
        except AttributeError:
            continue

# Scrape the first 100 reviews
while len(reviews) < 100:
    scrape_page()
    try:
        next_button = driver.find_element(By.XPATH, "//a[@class='_1LKTO3']")
        next_button.click()
        time.sleep(2)
    except Exception as e:
        print("No more pages or unable to click next button:", e)
        break

# Create a DataFrame
df = pd.DataFrame(reviews[:100], columns=['Rating', 'Review Summary', 'Full Review'])

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the Flipkart website and search for "sneakers"
driver.get("https://www.flipkart.com/")
time.sleep(5)

# Close the login popup if it appears
try:
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), '✕')]"))
    )
    close_button.click()
except Exception as e:
    print("Login popup did not appear or could not be closed:", e)

# Enter "sneakers" in the search field and search
search_field = driver.find_element(By.NAME, "q")
search_field.send_keys("sneakers")
search_field.send_keys(Keys.RETURN)

# Wait for the search results to load
time.sleep(5)

sneakers = []

# Function to scrape data from the current page
def scrape_page():
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    sneaker_blocks = soup.find_all('div', {'class': '_1AtVbE col-12-12'})
    
    for block in sneaker_blocks:
        try:
            brand = block.find('div', {'class': '_2WkVRV'}).text
            description = block.find('a', {'class': 'IRpwTa'}).text
            price = block.find('div', {'class': '_30jeq3'}).text
            sneakers.append([brand, description, price])
        except AttributeError:
            continue

# Scrape the first 100 sneakers
while len(sneakers) < 100:
    scrape_page()
    try:
        next_button = driver.find_element(By.XPATH, "//a[@class='_1LKTO3']")
        next_button.click()
        time.sleep(2)
    except Exception as e:
        print("No more pages or unable to click next button:", e)
        break

# Create a DataFrame
df = pd.DataFrame(sneakers[:100], columns=['Brand', 'Product Description', 'Price'])

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the Amazon website
driver.get("https://www.amazon.in/")

# Find the search field and enter "Laptop"
search_field = driver.find_element(By.ID, "twotabsearchtextbox")
search_field.send_keys("Laptop")

# Click the search icon
search_button = driver.find_element(By.ID, "nav-search-submit-button")
search_button.click()

# Wait for the search results to load
time.sleep(5)

# Hover over the "CPU Type" filter and click "Intel Core i7"
cpu_filter = driver.find_element(By.XPATH, "//span[text()='CPU Type']/ancestor::a")
ActionChains(driver).move_to_element(cpu_filter).perform()
time.sleep(2)
intel_core_i7_filter = driver.find_element(By.XPATH, "//span[text()='Intel Core i7']")
intel_core_i7_filter.click()

# Wait for the filter to apply
time.sleep(5)

# Scrape the data for the first 10 laptops
laptops = []
for i in range(1, 11):
    try:
        laptop_title = driver.find_element(By.XPATH, f"(//span[@class='a-size-medium a-color-base a-text-normal'])[{i}]").text
        laptop_ratings = driver.find_element(By.XPATH, f"(//span[@class='a-icon-alt'])[{i}]").text
        laptop_price = driver.find_element(By.XPATH, f"(//span[@class='a-price-whole'])[{i}]").text
        laptops.append([laptop_title, laptop_ratings, laptop_price])
    except Exception as e:
        print(f"Error while scraping laptop {i}: {e}")

# Create a DataFrame
df = pd.DataFrame(laptops, columns=['Title', 'Ratings', 'Price'])

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Open the AzQuotes website
driver.get("https://www.azquotes.com/")

# Click on "Top Quotes"
top_quotes_link = driver.find_element(By.XPATH, "//a[text()='Top Quotes']")
top_quotes_link.click()

# Wait for the page to load
time.sleep(5)

# Scroll down to load more quotes (repeat this multiple times to load more quotes)
for _ in range(10):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Scrape the data for the top 1000 quotes
quotes = []
quote_elements = driver.find_elements(By.CLASS_NAME, "title")
author_elements = driver.find_elements(By.CLASS_NAME, "author")
type_elements = driver.find_elements(By.CLASS_NAME, "kw-box")

for i in range(len(quote_elements)):
    quote = quote_elements[i].text
    author = author_elements[i].text
    quote_type = type_elements[i].text
    quotes.append([quote, author, quote_type])

# Create a DataFrame
df = pd.DataFrame(quotes, columns=['Quote', 'Author', 'Type of Quote'])

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```
